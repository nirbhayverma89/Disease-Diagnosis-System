```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv(r'/content/sample_data/dataset.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.fillna("")
df['Symptom']=""
for i in range(1,18):
    df['s']=df["Symptom_{}".format(i)]
    df['Symptom']=df['Symptom']+df['s']
```

```python
df.head(10)
```

```python
for i in range(1,18):
    df=df.drop("Symptom_{}".format(i),axis=1)
df=df.drop("s",axis=1)
```

```python
df.head()
```

```python
df[0:1].values
```

```python
df['Disease'].value_counts()
```

```python
X=df['Symptom']
y=df['Disease']
```

```python
X.head(10)
```

```python
y.head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.25,shuffle=True,random_state=44)
```

```python
y_train.head(10)
```

```python
print('Training Data Shape:', X_train.shape)
print('Testing Data Shape: ', X_test.shape)
```

```python
y_train.value_counts()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()

X_train_tfidf = vectorizer.fit_transform(X_train)
X_train_tfidf.shape
```

```python
pd.DataFrame(X_train_tfidf)[0]
```

```python
from sklearn.svm import LinearSVC
clf = LinearSVC()
clf.fit(X_train_tfidf,y_train)
```

```python
from sklearn.pipeline import Pipeline
text_clf = Pipeline([('tfidf', TfidfVectorizer()),
                     ('clf', LinearSVC()),])


text_clf.fit(X_train, y_train)


predictions = text_clf.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(y_test,predictions))
```

```python
print(metrics.classification_report(y_test,predictions))
```

```python
print(metrics.accuracy_score(y_test,predictions))
```

```python
ds=pd.read_csv(r'/content/sample_data/symptom_Description.csv')
ds.head(10)
```

```python
ds.isnull().sum()
```

```python
ds.index=ds['Disease']
ds=ds.drop('Disease',axis=1)
ds.head(10)
```

```python
pr=pd.read_csv(r'/content/sample_data/symptom_precaution.csv')
pr.head(10)
```

```python
pr.isnull().sum()
```

```python
pr = pr.fillna("")
pr.head(10)
```

```python
pr['precautions']=""
pr['punc']=', '
for i in range(1,5):
    pr['s']=pr["Precaution_{}".format(i)]+pr['punc']
    pr['precautions']=pr['precautions']+pr['s']
pr.head(10)
```

```python
for i in range(1,5):
    pr=pr.drop("Precaution_{}".format(i),axis=1)
pr.head(10)
```

```python
pr=pr.drop(['s','punc'],axis=1)
pr.head(10)
```

```python
pr.index=pr['Disease']
pr=pr.drop('Disease',axis=1)
pr.head()
```

```python
def predict_top5_diseases(user_symptoms):
    # Get decision function scores
    decision_scores = text_clf.decision_function([user_symptoms])

    # Get top 5 indices with highest scores
    top5_indices = np.argsort(decision_scores[0])[-5:][::-1]

    # Map indices to diseases
    possible_diseases = text_clf.classes_[top5_indices]

    # Get decision scores for the top 5 diseases
    scores = decision_scores[0][top5_indices]

    # Apply logarithmic scaling to adjust small scores
    log_scores = np.log1p(scores)  # Log(1 + score) to avoid negative infinity for small/zero scores

    # Normalize using Softmax function to ensure that the probabilities sum to 1
    exp_scores = np.exp(log_scores - np.max(log_scores))  # Stability fix by subtracting max
    probabilities = exp_scores / np.sum(exp_scores)

    # Prepare results
    results = []
    for disease, probability in zip(possible_diseases, probabilities):
        description = ds.loc[disease, 'Description'] if disease in ds.index else "No description available"
        precautions = pr.loc[disease, 'precautions'] if disease in pr.index else "No precautions available"
        results.append({
            "Disease": disease,
            "Probability": round(probability * 100, 2),  # Convert to percentage
            "Description": description,
            "Precautions": precautions
        })

    return results
user_input_symptoms = "fever, headache, muscle_wasting"
top5_predictions = predict_top5_diseases(user_input_symptoms)
```

```python
print("\nTop 5 Prediction Results:")
for prediction in top5_predictions:
    print(f"\nDisease: {prediction['Disease']}")
    print(f"Probability: {prediction['Probability']}%")
    print(f"Description: {prediction['Description']}")
    print(f"Precautions: {prediction['Precautions']}")
```

```python
import matplotlib.pyplot as plt

# Print the top 5 predictions as before
print("\nTop 5 Prediction Results:")
diseases = []
probabilities = []

for prediction in top5_predictions:
    disease = prediction['Disease']
    probability = prediction['Probability']
    print(f"\nDisease: {disease}")
    print(f"Probability: {probability}%")

    diseases.append(disease)
    probabilities.append(probability)

# Create a bar plot with switched axes
plt.figure(figsize=(10, 6))
plt.bar(diseases, probabilities, color='skyblue')
plt.ylabel('Probability (%)')
plt.title('Top 5 Disease Predictions')

# Adding the percentage values inside the bars
for i in range(len(diseases)):
    plt.text(i, probabilities[i] + 2, f'{probabilities[i]:.2f}%', ha='center', va='bottom', color='black', fontweight='bold')

plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```
